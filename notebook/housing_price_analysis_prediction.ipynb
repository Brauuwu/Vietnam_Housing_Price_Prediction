### 🛠️ Preprocessing - Tiền xử lý dữ liệu

---

#### 1️⃣ Tạo trạng thái lưu đồ thị

```python
# Savestates graph
plot_all = {}
```

#### 1️⃣ Sao lưu dữ liệu

```python
org_hb = df.copy()
```

#### 2️⃣ Loại bỏ các thuộc tính không cần thiết

```python
org_hb = org_hb.drop(columns = ['id', 'detail_url', 'title'])
```

#### 3️⃣ Phân rã Location và Timeline

```python
# Tách Location thành 2 cột thuộc tính District và Province
org_hb[['district', 'province']] = org_hb['location'].str.split(',', n = 1, expand=True)

# Loại bỏ khoảng trắng sau khi tách
org_hb['district'] = org_hb['district'].str.strip()
org_hb['province'] = org_hb['province'].str.strip()

# Loại bỏ cột Location
org_hb = org_hb.drop(columns = 'location')
```

```python
# Tạo dạng thời điểm ngày tham chiếu
reference_date = datetime(2025, 9, 21)

# Chuyển đổi từ Timeline (Hours) sang Datetime
org_hb['date'] = org_hb['timeline_hours'].apply(lambda h: reference_date - timedelta(hours=h))

# Trích xuất ngày, tháng, năm tạo thành cột thuộc tính
org_hb['day'] = org_hb['date'].dt.day
org_hb['month'] = org_hb['date'].dt.month
org_hb['year'] = org_hb['date'].dt.year

# Loại bỏ các cột Timeline (Hours), Date
org_hb = org_hb.drop(columns=['timeline_hours', 'date'])
```

#### 4️⃣ Loại bỏ bản ghi chứa các Thuộc tính không hợp lệ và các ngoại lai

```python
# Đếm số lượng đặc trưng cấu trúc bị thiếu trên mỗi dòng
missing_struct_count = org_hb[['bedrooms', 'bathrooms', 'floors']].isna().sum(axis=1)

# 1. Điều kiện: Thiếu cả 3 đặc trưng cấu trúc
mask_all_struct_missing = missing_struct_count == 3

# 2. Điều kiện: Thiếu diện tích VÀ thiếu >= 2 đặc trưng cấu trúc
mask_drop_area = org_hb['area_m2'].isna() & (missing_struct_count >= 2)

# 3. Điều kiện: Thiếu giá nhà VÀ thiếu >= 2 đặc trưng cấu trúc
mask_drop_price = org_hb['price_million_vnd'].isna() & (missing_struct_count >= 2)

# Tổng hợp tất cả các dòng cần loại bỏ
mask_to_drop = mask_all_struct_missing | mask_drop_area | mask_drop_price

# Lọc giữ lại các dòng không thỏa mãn điều kiện loại bỏ
org_hb_filt = org_hb.loc[~mask_to_drop].copy()
```

```python
# Sao lưu dữ liệu và loại bỏ bản ghi chưa price ảo
df = org_hb_filt[org_hb_filt['price_million_vnd'] > 0].copy()

# Tạo ngưỡng phân vị Outlier
floor_cutoff = df['floors'].quantile(0.995)
area_cutoff = df['area_m2'].quantile(0.98)
bed_cutoff = df['bedrooms'].quantile(0.99)
bath_cutoff = df['bathrooms'].quantile(0.99)

df = df.loc[
    ((df['floors'] <= floor_cutoff) | (df['floors'].isna())) &
    ((df['area_m2'] <= area_cutoff) | (df['area_m2'].isna())) &
    ((df['bedrooms'] <= bed_cutoff) | (df['bedrooms'].isna())) &
    ((df['bathrooms'] <= bath_cutoff) | (df['bathrooms'].isna()))
]

# Sao lưu
cleaned_hb = df
```

#### 5️⃣ Phân khúc giá thị trường bất động sản

```python
# Tính toán phân vị
q1 = cleaned_hb['price_million_vnd'].quantile(0.25)
q3 = cleaned_hb['price_million_vnd'].quantile(0.75)

def price_segmentation(price):
    if price <= q1:
        return 'Low'
    elif price >= q3:
        return 'High'
    else:
        return 'Mid'

cleaned_hb['price_segment'] = cleaned_hb['price_million_vnd'].apply(price_segmentation)

# Kiểm tra phân phối
print(cleaned_hb['price_segment'].value_counts())
```

#### 6️⃣ Chuẩn hóa danh sách các tỉnh thành sau khi sáp nhập

```python
# Trước khi sáp nhập
south_sample = ['Hồ Chí Minh', 'Bình Dương', 'Bà Rịa Vũng Tàu', 'Đồng Nai', 'Bình Phước', 'Cần Thơ', 'Sóc Trăng', 'Hậu Giang']
central_coast_sample = ['Đà Nẵng', 'Huế', 'Quảng Nam', 'Bình Định', 'Gia Lai']
north_sample = ['Hà Nội', 'Bắc Ninh', 'Bắc Giang', 'Hưng Yên', 'Thái Bình', 'Hải Phòng', 'Hải Dương']

# Sau khi sáp nhập
province_mapping = {
    # Miền Bắc
    'Hà Nội' : ['Hà Nội'],
    'Tuyên Quang': ['Tuyên Quang', 'Hà Giang'],
    'Lai Châu' : ['Lai Châu'],
    'Điện Biên' : ['Điện Biên'],
    'Sơn La' : ['Sơn La'],
    'Lạng Sơn' : ['Lạng Sơn'],
    'Cao Bằng' : ['Cao Bằng'],
    'Quảng Ninh' : ['Quảng Ninh'],
    'Lào Cai': ['Lào Cai', 'Yên Bái'],
    'Thái Nguyên': ['Thái Nguyên', 'Bắc Kạn'],
    'Phú Thọ': ['Phú Thọ', 'Vĩnh Phúc', 'Hòa Bình'],
    'Bắc Ninh': ['Bắc Ninh', 'Bắc Giang'],
    'Hưng Yên': ['Hưng Yên', 'Thái Bình'],
    'Hải Phòng': ['Hải Phòng', 'Hải Dương'],
    'Ninh Bình': ['Ninh Bình', 'Hà Nam', 'Nam Định'],

    # Miền Trung
    'Thanh Hoá' : ['Thanh Hoá'],
    'Nghệ An' : ['Nghệ An'],
    'Hà Tĩnh' : ['Hà Tĩnh'],
    'Huế' : ['Huế'],
    'Quảng Trị': ['Quảng Trị', 'Quảng Bình'],
    'Đà Nẵng': ['Đà Nẵng', 'Quảng Nam'],
    'Quảng Ngãi': ['Quảng Ngãi', 'Kon Tum'],
    'Gia Lai': ['Gia Lai', 'Bình Định'],
    'Đắk Lắk': ['Đắk Lắk', 'Phú Yên'],
    'Khánh Hòa': ['Khánh Hòa', 'Ninh Thuận'],
    'Lâm Đồng': ['Lâm Đồng', 'Đắk Nông', 'Bình Thuận'],

    # Miền Nam
    'Đồng Nai': ['Đồng Nai', 'Bình Phước'],
    'Hồ Chí Minh': ['Hồ Chí Minh', 'Bà Rịa Vũng Tàu', 'Bình Dương'],
    'Tây Ninh': ['Tây Ninh', 'Long An'],
    'Đồng Tháp': ['Đồng Tháp', 'Tiên Giang'],
    'Vĩnh Long': ['Vĩnh Long', 'Bến Tre', 'Trà Vinh'],
    'Cần Thơ': ['Cần Thơ', 'Sóc Trăng', 'Hậu Giang'],
    'Cà Mau': ['Cà Mau', 'Bạc Liêu'],
    'An Giang': ['An Giang', 'Kiên Giang']
}

# Hàm ánh xạ tỉnh cũ và mới
def map_old_new_provinces(province):
    # Kiểm tra các tỉnh không hợp lệ
    if pd.isna(province):
        return 'Other'

    prov = province.strip().lower()

    for new_province, old_list in province_mapping.items():
        for old_name in old_list:
            if prov == old_name.lower():
                return new_province
    return 'Other'

# Apply mapping to create a new column
cleaned_hb['province_after_emergence'] = cleaned_hb['province'].apply(map_old_new_provinces)
```

```python
# Remove outliers using IQR
Q1 = hb_price["price_million_vnd"].quantile(0.25)
Q3 = hb_price["price_million_vnd"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

hb_price_clean = hb_price[
    (hb_price["price_million_vnd"] >= lower_bound) &
    (hb_price["price_million_vnd"] <= upper_bound)
]

# Plot distribution after removing outliers
fig, ax = plt.subplots(figsize=(8,5))

sns.violinplot(data=hb_price_clean, y="price_million_vnd", ax=ax)

ax.set_title("Distribution of House Prices - Outliers Excluded")
ax.set_ylabel("Price (Million VND)")

plt.tight_layout()
plot_all["hp_price_dist_t1"] = fig
plt.show()
```

Mặc dù có nhiều ưu điểm, tập dữ liệu này vẫn tồn tại một số thách thức:

* **Dữ liệu thiếu / Null (Missing / Null Values)**
    * Thường gặp ở các đặc trưng số học cốt lõi như: Số phòng ngủ, Số phòng tắm, Số tầng. Điều này có thể gây sai lệch (bias) cho hiệu suất của mô hình nếu không được xử lý cẩn thận.
* **Các thuộc tính phức hợp**
    * Một số đặc trưng là sự kết hợp của nhiều thông tin về đơn vị hành chính và các thuộc tính liên quan đến thời gian.
    * Đòi hỏi phải phân tách thành các thành phần đa lớp để có thể diễn giải dữ liệu một cách có ý nghĩa.
* **Phân phối địa lý mất cân bằng**
    * Các tin đăng tập trung quá dày đặc ở các tỉnh/thành phố lớn.
    * Các tỉnh thành nhỏ hơn có rất ít dữ liệu đại diện.
    * Làm tăng nguy cơ mô hình bị overfitting và có khả năng khái quát hóa kém.
* **Chuẩn hóa đơn vị & Thang giá (Price Scale & Unit Normalization)**
    * Giá cả hiện đang được ghi nhận ở đơn vị triệu VNĐ.
    * Cần chuyển đổi sang đơn vị tỷ VNĐ để dữ liệu trở nên dễ diễn giải hơn và giúp ổn định các dự đoán của mô hình.
